```python
import signac
import os
import matplotlib.pyplot as plt
import numpy as np
project = signac.get_project()
```

```python
case, voltage, seed = "acn_litfsi", 2, 2
for job in project.find_jobs({'case': case, 'seed': seed, 'voltage': voltage}):
    with job:
        print(job.id)
        ### plot time vs charge
        data = np.load('charge.npy')
        pos_charge = np.sum(data[:int(len(data[0])/2)], axis= 1)
        time = np.arange(0, len(pos_charge), 1)
        plt.plot(time*2, pos_charge)
        plt.xlabel('Time')
        plt.ylabel('Accumulated charge')
```
